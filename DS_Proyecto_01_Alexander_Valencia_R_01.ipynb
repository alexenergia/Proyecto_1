```python
#B Estudia la distribución de esa variable: surface_total
#C.Lo mismo ocurre con valores de superficie total muy chico
sns.set()
sns.boxplot(y = "surface_total", data = new_data_fr)
print("Hay 4 grupos de datos, uno por debajo de 20.000 m2, uno entre 40 y 60 mil m2 y dos grupos outliners por encima de los 100.000m2")
```

```python
#Para eliminar los outliers (valores muy altos o muy pequeóños, se utiliza los rangos por Cuartiles asi:
q1,q3 = new_data_fr['surface_total'].quantile(0.25),new_data_fr['surface_total'].quantile(0.75)
IQR = q3-q1
superior= q3 + (1.5 * IQR) #se define el rango intercuatílico superior
inferior = q1 - (1.5 * IQR)  #se define el rango intercuatílico inferior

#Con esta rutina, se eliminan los outliers con base en el rango intercuartil:

new_data_fr = new_data_fr[(new_data_fr.surface_total > inferior) & (new_data_fr.surface_total < superior)]
sns.boxplot(x="property_type", y="surface_total", data =new_data_fr).set(title="Distribución",xlabel="Tipo",ylabel="superficie total")
plt.show()
print("Al eliminar los outliers, el dataframe queda asi:", new_data_fr.shape)
new_data_fr.describe()
```

```python
#DLas propiedades no pueden tener surface_covered mayor a surface_total. Si eso sucede, debes filtrar esas instancias.

new_data_fr = new_data_fr[new_data_fr.surface_total > new_data_fr.surface_covered ]
sns.boxplot(x="property_type", y="surface_total", data =new_data_fr).set(title="Distribución",xlabel="Tipo",ylabel="superficie total")
plt.show()
print("Al eliminar las propiedades con superficie incoherente, el dataframe queda asi:", new_data_fr.shape)
new_data_fr.describe()
```

```python
#El rango de precios que toman las propiedades es muy amplio.  Puede ser útil un boxplot para determinar un rango razonable.

#Para eliminar los outliers (valores muy altos o muy pequeños del precio, se utiliza los rangos por Cuartiles asi:
q1,q3 = new_data_fr['price'].quantile(0.25),new_data_fr['price'].quantile(0.75)
IQR = q3-q1
superior= q3 + (1.5 * IQR) #se define el rango intercuatílico superior
inferior = q1 - (1.5 * IQR)  #se define el rango intercuatílico inferior
new_data_fr = new_data_fr[(new_data_fr.price > inferior) & (new_data_fr.price < superior)]
sns.boxplot(x="property_type", y="price", data =new_data_fr).set(title="Distribución",xlabel="Tipo",ylabel="Precio")
plt.show()

plt.figure(figsize=(15,5))
sns.displot(new_data_fr['price']).set(title = "Distribución de los precios" )
plt.show()

print("Al eliminar los outliers, el dataframe queda asi:", new_data_fr.shape)
new_data_fr.describe()
```

```python
#D Una vez filtrado el dataset, puedes utilizar la función pairplot de Seaborn.
#Distribuciones y relaciones de a pares de las variables: rooms, bedrooms, bathrooms, surface_total, surface_covered, price para cada tipo de propiedad. Para ello, ten en cuenta:
pares1= ['surface_total','surface_covered','price']
sns.pairplot(data=new_data_fr[pares1], corner= True)
plt.show()
print("Se identifican relaciones significativas entre el precio y el área total y área cubierta, la relación entre áreas está limitada por el filtro realizado en la etapa previa")
```

```python
#Ahora a entrenar el modelo.
#En esta sección, debes entrenar dos modelos de Machine Learning - uno de vecinos más cercanos y otro de árboles de decisión -  para predecir el precio de las propiedades tipo `Departamento`, `PH` y `Casa`  
#en la Ciudad Autónoma de Buenos Aires (`Capital Federal`).
#1 Generar un modelo benchmark, en este caso una regresión lineal entre price versus surface_covered / surface_total / rooms como Features y Price como target

# X= Surface covered-------------------------------------------------------------------
#Regresion lineal para el modelo X= Surface covered / y = Price.
X,y = new_data[['surface_covered']].values, new_data[['price']].values
from sklearn.linear_model import LinearRegression
LRsimple = LinearRegression()
LRsimple.fit(X,y)
yfit = LRsimple.predict(X)

plt.scatter(X,y, color ="blue")
plt.plot(X,yfit, color ="red", linestyle = 'dashed')
plt.ylabel('Price')
plt.xlabel('Surface Covered')
plt.show()

#Ahora el RMSE para verficar la adecuación o bondad de ajuste del modelo X= Surface covered / y = Price.

#Calculando el la Raiz del Error Cuadrático Medio, RMSE
from sklearn.metrics import mean_squared_error
RMSE = np.sqrt(mean_squared_error(y,yfit))               
print("El RMSE de la regresión lineal es:", RMSE )

#Calculando el coeficiente de ajuste R2
R2 = LRsimple.score(X,y)
print ("El coeficiente de determinación R2:", R2)

print("Es necesario comparar la bondad del ajuste respecto a las otras variables")

# X= Surface Total-----------------------------------------------------------------------------------------------
#Regresion lineal para el modelo X= Surface Total / y = Price.
X,y = new_data[['surface_total']].values, new_data[['price']].values
from sklearn.linear_model import LinearRegression
LRsimple = LinearRegression()
LRsimple.fit(X,y)
yfit = LRsimple.predict(X)

plt.scatter(X,y, color ="green")
plt.plot(X,yfit, color ="red", linestyle = 'dashed')
plt.ylabel('Price')
plt.xlabel('Surface Total')
plt.show()

#Ahora el RMSE para verficar la adecuación o bondad de ajuste del modelo X= Surface Total / y = Price.

#Calculando el Error cuadrático medio
from sklearn.metrics import mean_squared_error
RMSE = np.sqrt(mean_squared_error(y,yfit))               
print("El RMSE de la regresión lineal es:", RMSE )

#Calculando el coeficiente de ajuste R2
R2 = LRsimple.score(X,y)
print ("El coeficiente de determinación R2:", R2)
print("Es necesario comparar la bondad del ajuste respecto a las otras variables")

# X= rooms -----------------------------------------------------------------

#Regresion lineal para el modelo X= rooms / y = Price.
X,y = new_data[['rooms']].values, new_data[['price']].values
from sklearn.linear_model import LinearRegression
LRsimple = LinearRegression()
LRsimple.fit(X,y)
yfit = LRsimple.predict(X)

plt.scatter(X,y, color ="yellow")
plt.plot(X,yfit, color ="red", linestyle = 'dashed')
plt.ylabel('Price')
plt.xlabel('Rooms')
plt.show()

#Ahora el RMSE para verficar la adecuación o bondad de ajuste del modelo X= Surface Total / y = Price.

#Calculando el Error cuadrático medio
from sklearn.metrics import mean_squared_error
RMSE = np.sqrt(mean_squared_error(y,yfit))               
print("El RMSE de la regresión lineal es:", RMSE )

#Calculando el coeficiente de ajuste R2
R2 = LRsimple.score(X,y)
print ("El coeficiente de determinación R2:", R2)

```

```python
#En esta sección se realizan los entrenamientos para la Regresion Lineal, KNN vecinos y Árboles:

#PARA REGRESION LINEAL:

#1) Partir el 80/20 de los datos para entrenar y predecir usando test_split
from sklearn.model_selection import train_test_split
X, y = X,y = new_data.drop(['price','surface_total'], axis =1).values, new_data[['price']].values
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.8, test_size = 0.2, random_state = 42) # The Hitchhiker's Guide to the Galaxy - 42 is the answer.
print(X_train.shape,  y_train.shape, X_test.shape,  y_test.shape)

#Regresion lineal para el modelo X= Surface Total / y = Price.  80% de los datos
modelo_lineal = LinearRegression() # Se define el tipo de modelo
modelo_lineal.fit(X_train, y_train) # Se entrena el modelo con el 80% de la data

from sklearn.metrics import mean_squared_error #Evaluar modelos de entrenamiento y prediccion
y_train_pred = modelo_lineal.predict(X_train) 
y_test_pred = modelo_lineal.predict(X_test) 


#-----------------------------------
RMSE_train = np.sqrt(mean_squared_error(y_train,y_train_pred)) # Se evalúa el modelo de entrenamiento              
print("El RMSE del modelo de regresión lineal en entreamiento es:", RMSE_train )
#Calculando el coeficiente de ajuste R2 para entrenamiento
R2_train = modelo_lineal.score(X_train,y_train)
print ("El coeficiente de determinación R2 para entrenamiento:", R2_train)

#-----------------------------------

RMSE_test = np.sqrt(mean_squared_error(y_test,y_test_pred)) # Se evalúa el modelo de predicción              
print("El RMSE del modelo de regresión lineal en predicción es:", RMSE_test )
#Calculando el coeficiente de ajuste R2 para predicción
R2_test = modelo_lineal.score(X_test,y_test)
print ("El coeficiente de determinación R2 para predicción:", R2_test)
print("El modelo de predicción obtuvo un mayor RMSE pero la bondad del ajuste cayó casí en un 5% respecto al de entrenamiento")

#finalmente una gráfica para comparar ambos modelos

fig,(ax1,ax2) = plt.subplots(1,2)
ax1.scatter(y_test, y_test_pred, color="green")
plt.title('Prueba', fontsize = 11)
ax2.scatter(y_train, y_train_pred, color="blue")
plt.title('Entrenamiento', fontsize = 11)
```

```python
#PARA ARBOLES : Se evaluarán los hiperparámetros que minimicen el RMSE para árboles

#Evaluar comportamiento de la profundidad del árbol
from sklearn.tree import DecisionTreeRegressor

lista_tree_train=[]
lista_tree_test=[]
n_tree = [1,2,3,4,5,6,7,10,15,20]

for i in n_tree:
    tree=DecisionTreeRegressor(max_depth = i)
    
    tree.fit(X_train,y_train)
    
    y_train_pred = tree.predict(X_train)
    rmse_train_tree = np.sqrt(mean_squared_error(y_train,y_train_pred))
    
    y_test_pred =tree.predict(X_test)
    rmse_test_tree = np.sqrt(mean_squared_error(y_test, y_test_pred))
    
    lista_tree_train.append(rmse_train_tree)
    lista_tree_test.append(rmse_test_tree)
```

```python
#Ahora graficar la evolución de los níveles de profundidad.
plt.plot(n_tree,lista_tree_train,"o-",label ="Entrenamiento")
plt.plot(n_tree, lista_tree_test,"-o",label ="Prueba")
plt.legend()
plt.xlabel("Profundidad del arbol")
plt.ylabel("RMSE")
```

```python
print ("Conclusión: La validación de la curva de aprendizaje, indica que el RMSE mínimo del arbol se dá para profundidad de MAX_DEPTH=10, luego el RMSE se estabiliza.")
```

```python
#PARA KNN VECINOS : Se evaluarán los hiperparámetros que minimicen el RMSE para KNN VECINOS
#Evaluar comportamiento de de la cantidad de fronteras en knn vecinos
from sklearn.neighbors import KNeighborsRegressor

lista_knn_train=[]
lista_knn_test=[]
k_vecinos = [1,2,3,4,5,6,7,10,15,20]

for i in k_vecinos:
    knn=KNeighborsRegressor(n_neighbors = i)
    
    knn.fit(X_train,y_train)
    
    y_train_pred = knn.predict(X_train)
    rmse_train_knn = np.sqrt(mean_squared_error(y_train,y_train_pred))
    
    y_test_pred =knn.predict(X_test)
    rmse_test_knn = np.sqrt(mean_squared_error(y_test, y_test_pred))
    
    lista_knn_train.append(rmse_train_knn)
    lista_knn_test.append(rmse_test_knn)
```

```python
#Ahora graficar la evolución de los níveles de profundidad.
plt.plot(k_vecinos,lista_knn_train,"o-",label ="Entrenamiento")
plt.plot(k_vecinos, lista_knn_test,"-o",label ="Prueba")
plt.legend()
plt.xlabel("Parámetro de fronteras")
plt.ylabel("RMSE")
```

```python
print ("Conclusión: La validación de la curva de aprendizaje, indica que el RMSE mínimo para KNN vecinos se dá para  # NEIGHBORS=3, luego el RMSE se estabiliza.")
```
